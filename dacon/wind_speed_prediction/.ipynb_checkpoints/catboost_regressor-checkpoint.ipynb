```python
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import OneHotEncoder

# train.csv  
train_data = pd.read_csv('train.csv')

#   
train_data = train_data.fillna(train_data.mean())

# ' '     
encoder = OneHotEncoder(sparse=False)
time_encoded = encoder.fit_transform(train_data[[' ']])
time_encoded_df = pd.DataFrame(time_encoded, columns=encoder.get_feature_names([' ']))
train_data = pd.concat([train_data, time_encoded_df], axis=1).drop([' '], axis=1)

#   ( ) ( ) .
X_train = train_data.drop(['ID', ' (m/s)'], axis=1)  #  
y_train = train_data[' (m/s)']  #   ()

# CatBoost       .
model = CatBoostRegressor(random_seed=42)
model.fit(X_train, y_train)

# test.csv  
test_data = pd.read_csv('test.csv')

#   
test_data = test_data.fillna(test_data.mean())

# ' '     
time_encoded = encoder.transform(test_data[[' ']])
time_encoded_df = pd.DataFrame(time_encoded, columns=encoder.get_feature_names([' ']))
test_data = pd.concat([test_data, time_encoded_df], axis=1).drop([' '], axis=1)

#     .
X_test = test_data.drop('ID', axis=1)  #   
y_pred = model.predict(X_test)  #    

# Submit / 
submission = pd.read_csv('./sample_submission.csv')
submission[' (m/s)'] = y_pred

#   submission.csv   .
submission.to_csv('submission.csv', index=False)

print("  .  submission.csv .")

# Cross Validation   RMSE 
from sklearn.model_selection import cross_val_score
import numpy as np
# Cross Validation   MAE 
mae_scores = -cross_val_score(model, X_train, y_train, scoring='neg_mean_absolute_error', cv=5)
print(" MAE with Ensemble:", np.mean(mae_scores))
```
